```python
# ============================================
# Problem 1: User Login Check
# ============================================

# Given credentials
username = "admin"
password = "1234"

# Input from user (optional: you can comment input and use fixed values)
input_username = input("Enter username: ")
input_password = input("Enter password: ")

# Check login
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# ============================================
# Problem 2: Pass / Fail Analyzer
# ============================================

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Analyze marks
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print result
print("\nPass / Fail Analysis:")
print("Total Students:", len(marks))
print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

```python
# ============================================
# Problem 3: Simple Data Cleaner
# ============================================

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

# Clean names
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print cleaned list
print("\nCleaned Names:")
print(cleaned_names)
```

```python
# ============================================
# Problem 4: Message Length Analyzer
# ============================================

messages = ["Hi", "Welcome to the platform", "OK"]

print("\nMessage Length Analysis:")

for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")

    if length > 10:
        print("Flag: Long message")

```

```python
# ============================================
# Problem 5: Error Message Detector
# ============================================

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

# Count errors
for log in logs:
    if log == "ERROR":
        error_count += 1

# Print total error count
print("\nError Log Analysis:")
print("Total ERROR entries:", error_count)
```
